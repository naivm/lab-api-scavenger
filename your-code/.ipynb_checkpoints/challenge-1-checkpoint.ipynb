```python
import pandas as pd
import numpy as np
import requests
import json
from pandas.io.json import json_normalize
```

```python
url = "https://api.github.com/repos/ironhack-datalabs/datalis0819/forks"
response_repos = requests.get(url)

response_repos.content
```

```python
norm_json = json_normalize(response_repos.json())
ds_normalized = pd.DataFrame(norm_json)
#ds_normalized["language"].describe()
ds_normalized.columns
```

```python
ds_language = np.array(ds_normalized["language"])
ds_unique = np.unique(ds_language)
print(ds_unique)
```
